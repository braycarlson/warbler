```python
import seaborn as sns

from datatype.dataset import Dataset
from matplotlib import pyplot as plt
```

```python
dataset = Dataset('exclude')
dataframe = dataset.load()
```

```python
folders = dataframe.folder.unique()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Exclusion per Individual')

sns.barplot(
    x=dataframe.folder.value_counts(),
    y=dataframe.folder.value_counts().index, 
    orient='h'
)

ax.set_xlabel('Exclusion')
ax.set_ylabel('Count')

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Most Excluded Recording')

x = dataframe.filename.value_counts()
y = dataframe.filename.value_counts().index

sns.barplot(
    x=x[:20],
    y=y[:20], 
    orient='h'
)

ax.set_xlabel('Exclusion')
ax.set_ylabel('Filename')

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Longest Segment Excluded')

x = dataframe.nlargest(20, 'duration').duration
y = dataframe.nlargest(20, 'duration').filename

sns.barplot(
    x=x,
    y=y,
    orient='h'
)

ax.set_xlabel('Duration (s)')
ax.set_ylabel('Filename')

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Most Frequently Excluded Sequence')

x = dataframe.sequence.value_counts()
y = dataframe.sequence.value_counts().index.astype('string')

sns.barplot(
    x=x,
    y=y,
    orient='h'
)

ax.set_xlabel('Count')
ax.set_ylabel('Sequence')

plt.show()
plt.close()
```
