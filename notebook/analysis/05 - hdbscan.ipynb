```python
import seaborn as sns

from constant import PROJECTION
from datatype.dataset import Dataset
from matplotlib import pyplot as plt
```

```python
dataset = Dataset('segment')
dataframe = dataset.load()
```

```python
folders = dataframe.folder.unique()
```

# 2D

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Amount of Segments per 2D Cluster')

sns.barplot(
    x=dataframe.hdbscan_label_2d.value_counts().index,
    y=dataframe.hdbscan_label_2d.value_counts(), 
    orient='v'
)

ax.set_xlabel('2D HDBScan Label')
ax.set_ylabel('Segment')

filename = 'hdbscan_amount_of_segments_per_2d_cluster.png'
path = PROJECTION.joinpath(filename)

plt.savefig(
    path,
    bbox_inches='tight',
    dpi=300,
    format='png'
)

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Amount of Segments per 2D Cluster (No noise)')

sns.barplot(
    x=dataframe[dataframe.hdbscan_label_2d > -1].hdbscan_label_2d.value_counts().index,
    y=dataframe[dataframe.hdbscan_label_2d > -1].hdbscan_label_2d.value_counts(), 
    orient='v'
)

ax.set_xlabel('2D HDBScan Label')
ax.set_ylabel('Segment')

filename = 'hdbscan_amount_of_segments_per_2d_cluster_no_noise.png'
path = PROJECTION.joinpath(filename)

plt.savefig(
    path,
    bbox_inches='tight',
    dpi=300,
    format='png'
)

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Mean Duration of Each 2D Cluster')

sns.barplot(
    x=dataframe.groupby('hdbscan_label_2d').duration.mean().index,
    y=dataframe.groupby('hdbscan_label_2d').duration.mean(), 
    orient='v'
)

ax.set_xlabel('2D HDBScan Label')
ax.set_ylabel('Duration (s)')

filename = 'hdbscan_mean_duration_of_each_2d_cluster.png'
path = PROJECTION.joinpath(filename)

plt.savefig(
    path,
    bbox_inches='tight',
    dpi=300,
    format='png'
)

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Mean Minimum Frequency of Each 2D Cluster')

sns.barplot(
    x=dataframe.groupby('hdbscan_label_2d').minimum.mean().index,
    y=dataframe.groupby('hdbscan_label_2d').minimum.mean(), 
    orient='v'
)

ax.set_xlabel('2D HDBScan Label')
ax.set_ylabel('Frequency (Hz)')

filename = 'hdbscan_mean_minimum_frequency_of_each_2d_cluster.png'
path = PROJECTION.joinpath(filename)

plt.savefig(
    path,
    bbox_inches='tight',
    dpi=300,
    format='png'
)

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Mean Maximum Frequency of Each 2D Cluster')

sns.barplot(
    x=dataframe.groupby('hdbscan_label_2d').maximum.mean().index,
    y=dataframe.groupby('hdbscan_label_2d').maximum.mean(), 
    orient='v'
)

ax.set_xlabel('2D HDBScan Label')
ax.set_ylabel('Frequency (Hz)')

filename = 'hdbscan_mean_maximum_frequency_of_each_2d_cluster.png'
path = PROJECTION.joinpath(filename)

plt.savefig(
    path,
    bbox_inches='tight',
    dpi=300,
    format='png'
)

plt.show()
plt.close()
```
