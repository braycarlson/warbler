```python
import seaborn as sns

from datatype.dataset import Dataset
from matplotlib import pyplot as plt
```

```python
dataset = Dataset('segment')
dataframe = dataset.load()
```

```python
folders = dataframe.folder.unique()
```

# 2D

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Amount of Segments per 2D Cluster')

sns.barplot(
    x=dataframe.fcm_label_2d.value_counts().index,
    y=dataframe.fcm_label_2d.value_counts(), 
    orient='v'
)

ax.set_xlabel('2D FCM Label')
ax.set_ylabel('Segment')

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Mean Duration of Each 2D Cluster')

sns.barplot(
    x=dataframe.groupby('fcm_label_2d').duration.mean().index,
    y=dataframe.groupby('fcm_label_2d').duration.mean(), 
    orient='v'
)

ax.set_xlabel('2D FCM Label')
ax.set_ylabel('Duration (s)')

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Mean Minimum Frequency of Each 2D Cluster')

sns.barplot(
    x=dataframe.groupby('fcm_label_2d').minimum.mean().index,
    y=dataframe.groupby('fcm_label_2d').minimum.mean(), 
    orient='v'
)

ax.set_xlabel('2D FCM Label')
ax.set_ylabel('Frequency (Hz)')

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(
    figsize=(18, 12)
)

ax.set_title('Mean Maximum Frequency of Each 2D Cluster')

sns.barplot(
    x=dataframe.groupby('fcm_label_2d').maximum.mean().index,
    y=dataframe.groupby('fcm_label_2d').maximum.mean(), 
    orient='v'
)

ax.set_xlabel('2D FCM Label')
ax.set_ylabel('Frequency (Hz)')

plt.show()
plt.close()
```
